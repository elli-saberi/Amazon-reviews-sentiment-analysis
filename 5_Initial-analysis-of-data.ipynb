<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
بارگذاری دیتاست اصلی شامل نظرات کاربران و امتیاز کلی هر محصول.
<br>
در این دیتاست به دو ستون زیر نیاز داریم:
<ul dir="rtl">
<li>
asin : این ستون شامل شناسه محصول است که به عنوان کلیدخارجی برای بدست آوردن عنوان و برند محصول از دیتاست دیگر استفاده میشود.
</li>
<li>
overall : مقدار امتیازی که به آن محصول در آن نظر داده شه است که عددی طبیعی از 1 تا 5 است.
</li>
</ul>
</font>
</p>

```python
train_data = pd.read_csv('/content/drive/MyDrive/project2/train_data.csv')
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
بارگذاری دیتاست کمکی شامل شناسه‌ی محصول (ASIN)، نام برند و عنوان محصول.
<br>
در این دیتاست به سه ستون زیر نیاز داریم:
<ul dir="rtl">
<li>
asin : شناسه محصول برای اتصال با دیتای اصلی
</li>
<li>
title : عنوان محصول
</li>
<li>
brand : برند و یا سازنده محصول
</li>
</ul>
</font>
</p>

```python
title_brand = pd.read_csv('/content/drive/MyDrive/project2/title_brand.csv')
```

```python
title_brand.head(10)
```

```python
title_brand.isna().sum()
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
در این بخش ابتدا محصولاتی که امتیاز ۵ دریافت کرده‌اند استخراج می‌شوند.
<br>
سپس با اطلاعات برند و عنوان محصول ادغام شده و بر اساس تعداد نظرات با امتیاز ۵ مرتب می‌شوند.
<br>
در نهایت، ۱۰ محصول برتر با بیشترین تعداد امتیاز ۵ در قالب یک جدول رنگی نمایش داده می‌شوند.
</font>
</p>

```python
product_counts = train_data[train_data['overall'] == 5].groupby('asin').size().reset_index(name='five_star_count')
merged = pd.merge(product_counts, title_brand, on='asin', how='left').dropna(subset=['brand', 'title'])

top10 = merged.sort_values('five_star_count', ascending=False).head(10)
top10_final = top10[['brand', 'title', 'five_star_count']].reset_index(drop=True)

print("Products with the most number of 5 starts: ")
display(top10_final.style.background_gradient(cmap='Greens'))
```
